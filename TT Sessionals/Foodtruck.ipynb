```python
# Name: Soham Samanta | Roll : 20051107 | CSE-21 | T&T Lab Sessional
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
from sklearn.linear_model import LinearRegression
from sklearn.model_selection import train_test_split
```

```python
data = pd.read_csv('Foodtruck.csv')
```

```python
# Check for missing values
print(data.isnull().sum())

# Split the dataset into features (X) and labels (y)
X = data.iloc[:, :-1].values
y = data.iloc[:, -1].values
```

```python
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.2, random_state=0)
```

```python
regressor = LinearRegression()
regressor.fit(X_train, y_train)
```

```python
#2
# Training set visualization
plt.scatter(X_train, y_train, color='red')
plt.plot(X_train, regressor.predict(X_train), color='blue')
plt.title('Profit vs Population (Training set)')
plt.xlabel('Population in 10000s')
plt.ylabel('Profit in $10,000s')
plt.show()

# Test set visualization
plt.scatter(X_test, y_test, color='red')
plt.plot(X_train, regressor.predict(X_train), color='blue')
plt.title('Profit vs Population (Test set)')
plt.xlabel('Population in 10000s')
plt.ylabel('Profit in $10,000s')
plt.show()
```

```python
#3
jaipur_population = 3.073 * 100  # Convert the population to the same unit as in the dataset
jaipur_profit = regressor.predict([[jaipur_population]])

print(f"Estimated profit for a food truck in Jaipur with a population of 3.073 million: ${jaipur_profit[0] * 10000:.2f}")
```
