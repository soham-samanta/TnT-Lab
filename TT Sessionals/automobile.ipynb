```python
# 2. Get the values from Price column into a numpy.ndarray
pricearr = df["price"].to_numpy()
print(pricearr)
```

```python
# 3. Calculate the Minimum Price, Maximum Price, Average Price, and Standard Deviation of Price
print(pricearr.max())
print(pricearr.min())
print(pricearr.mean())
print(pricearr.std())
```

```python
# 4. Make a pie chart for all car makers
car_makers = df['make'].value_counts()
plt.pie(car_makers, labels=car_makers.index, autopct='%1.1f%%', shadow=True)
plt.title("Car Makers Distribution")
plt.axis('equal')
plt.show()
```
